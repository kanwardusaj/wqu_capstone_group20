# ANALYSIS

### DATA AND METHODOLOGY


The project analysis daily and monthly adjusted close prices of stock market indices for 10 selected stocks (APPL, IBM, CNSL, CSL, KO, EBAY, DIS, HPQ, TGT and MSFT). Observation are taken for the period 1/01/2005 â€“ 01/01/2020 for which data are available. The analysis investigates the weak form of market efficiency using the selected stock prices from yahoo finance. It is to investigate whether the select stocks follows a random walk model or not. In effect, no arbitrage profits can be realized if the market is efficient. 

### Normality Test 

From code run, the monthly returns are negatively skewed in all selected stocks for the selected period. This indicates large negative returns (minimum extreme value) beyond -0.4 which are larger than the higher positive returns (maximum extreme value) a little beyond 0.3. The kurtosis is positive across the selected stocks which indicates leptokurtic distribution of monthly returns. This shows higher peaks than expected from normal distributions. Further, a QQ-Plot performed in the code clearly indicates that the distribution of monthly returns does not follow a normal distribution as the tail ends have outliers clearly spread further away from the normal distribution line shown in red in code implementation. MSFT has the highest return of 2.49% with Std Dev of 6.55% whiles CNSL has least return of -5.10% with a Std Dev of 10.05%

### Randomness Test

To analyze randomness of the monthly returns, autocorrelation plot of monthly returns was performed for the selected period. It is clear from the plot that, the null of the entire autocorrelation coefficients together equal to zero may be rejected. This indicates that historical returns can be used to correctly predict future returns of stocks. This further shows that the weak form of efficient market hypothesis does not hold entirely.
